# Convolutional Neural Network Code with MNIST

This is jupyter notebook for basic CNN and dense layer using `Tensorflow`. 
<br> Dataset is MNIST dataset.

```python
# Import libraries
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Import MNIST data
from tensorflow.examples.tutorials.mnist import input_data
mnist = input_data.read_data_sets('MNIST_data', validation_size=0)
```

## Parameters

```python
# Parameters 
img_size = 28
img_flat_size = img_size * img_size

# labels: 0 - 9
num_label = 10

# Parameters for optimizer
num_epoch = 10

learning_rate = 5e-4
epsilon = 1e-8

batch_size = 512

validation_ratio = 0.1
gpu_fraction = 0.4

# Convolution parameters [kernel size, kernel size, # of channels(input), # of channels (output)]
first_conv  = [3,3, 1,32]
second_conv = [3,3,32,64]
third_conv  = [3,3,64,128]

# Fully connected parameters [# of input data, # of output data]
first_fc = [4 * 4 * 128, 512]
second_fc = [512, num_label]
```

```python
# Plotting example image
img = mnist.train.images[0]
img_resize = img.reshape((img_size, img_size))
plt.imshow(img_resize, cmap = 'gray')
plt.show()
```

## Functions

```python
# Initialize weights and bias 
def conv_weight_variable(name, shape):
    return tf.get_variable(name, shape = shape, initializer = tf.contrib.layers.xavier_initializer_conv2d())

def weight_variable(name, shape):
    return tf.get_variable(name, shape=shape, initializer=tf.contrib.layers.xavier_initializer())

def bias_variable(name, shape):
    return tf.get_variable(name, shape=shape, initializer=tf.contrib.layers.xavier_initializer())

# Convolution and pooling
def conv2d(x,w, stride):
    return tf.nn.conv2d(x,w,strides=[1, stride, stride, 1], padding='SAME')

def max_pool_2x2(x):
    return tf.nn.max_pool(x, ksize=[1,2,2,1], strides=[1,2,2,1], padding='SAME')
```

## CNN Network

```python
# Network

# Reset graph
tf.reset_default_graph()

# Input 
x_image  = tf.placeholder(tf.float32, shape = [None, img_size, img_size, 1])
y_target = tf.placeholder(tf.float32, shape=[None, num_label])

# Keep Prob
keep_prob = tf.placeholder("float")

# Convolution variables
w_conv1 = conv_weight_variable('w_conv1', first_conv)
b_conv1 = bias_variable('b_conv1', [first_conv[3]])

w_conv2 = conv_weight_variable('w_conv2', second_conv)
b_conv2 = bias_variable('b_conv2', [second_conv[3]])

w_conv3 = conv_weight_variable('w_conv3', third_conv)
b_conv3 = bias_variable('b_conv3', [third_conv[3]])
    
# Densely connect layer variables 
w_fc1 = weight_variable('w_fc1', first_fc)
b_fc1 = bias_variable('b_fc1', first_fc[1])

w_fc2 = weight_variable('w_fc2', second_fc)
b_fc2 = bias_variable('b_fc2', second_fc[1])

# Conv layers
h_conv1 = tf.nn.relu(conv2d(x_image, w_conv1, 1) + b_conv1)
h_pool1 = max_pool_2x2(h_conv1)

h_conv2 = tf.nn.relu(conv2d(h_pool1, w_conv2, 1) + b_conv2)
h_pool2 = max_pool_2x2(h_conv2)

h_conv3 = tf.nn.relu(conv2d(h_pool2, w_conv3, 1) + b_conv3)
h_pool3 = max_pool_2x2(h_conv3)

# Flatten
h_flatten = tf.reshape(h_pool3, [-1, 4 * 4 * 128])

# Fully connected Layer
h_fc1 = tf.nn.relu(tf.matmul(h_flatten, w_fc1)+b_fc1)

# Dropout
keep_prob = tf.placeholder("float")
h_fc1_drop = tf.nn.dropout(h_fc1, keep_prob)

output = tf.nn.softmax(tf.matmul(h_fc1_drop, w_fc2)+b_fc2)
```

## Set Loss and Optimizer

```python
# Training 
Loss = tf.reduce_mean(tf.square(output - y_target))
optimizer = tf.train.AdamOptimizer(learning_rate = learning_rate, epsilon = epsilon).minimize(Loss)

correct_prediction = tf.equal(tf.argmax(y_target,1), tf.argmax(output,1))
accuracy = tf.reduce_mean(tf.cast(correct_prediction, tf.float32))
```

## Making Dataset

```python
# Dataset for train, test, validation

test_len = mnist.test.images.shape[0]
validation_len = int(test_len * validation_ratio)

train_x = mnist.train.images
test_x = mnist.test.images[validation_len : test_len, :]
validation_x = mnist.test.images[ : validation_len, :]

# Reshape
validation_x_resize = np.reshape(validation_x, (validation_x.shape[0], img_size, img_size, 1))

train_y = mnist.train.labels
test_y = mnist.test.labels[validation_len : test_len]
validation_y = mnist.test.labels[ : validation_len]

print("Training set: " + str(train_x.shape))
print("Testing set: " + str(test_x.shape))
print("Validation set: " + str(validation_x.shape))
```

## Create Session

```python
# Create Session
config = tf.ConfigProto()
config.gpu_options.per_process_gpu_memory_fraction = gpu_fraction

sess = tf.InteractiveSession(config=config)
sess.run(tf.global_variables_initializer())
```

## Training

```python
data = np.insert(train_x, img_flat_size, train_y, axis = 1)
len_data = data.shape[0]

# Training
for i in range(num_epoch):
    # Shuffle the data 
    np.random.shuffle(data)

    # Split data to input can label
    data_x = data[:, :img_flat_size]
    data_y = data[:, img_flat_size]

    # Make label data as one-hot
    data_y_onehot = np.zeros([data_y.shape[0], num_label])
    for j in range(data_y.shape[0]):
        data_y_onehot[j, int(data_y[j])] = 1
    
    data_y_onehot_val = np.zeros([validation_y.shape[0], num_label])
    for j in range(validation_y.shape[0]):
        data_y_onehot_val[j, int(validation_y[j])] = 1
        
    # Making mini-batch
    for j in range(0, len_data, batch_size):
        if j + batch_size < len_data:
            data_x_in = data_x[j : j + batch_size, :]
            data_y_in = data_y_onehot[j : j + batch_size, :]
        else:
            data_x_in = data_x[j : len_data, :]
            data_y_in = data_y_onehot[j : len_data, :]

        # Reshape input data as image
        data_x_in_resize = np.reshape(data_x_in, (data_x_in.shape[0], img_size, img_size, 1))

        # Run Optimizer!
        optimizer.run(feed_dict = {x_image: data_x_in_resize, y_target: data_y_in, keep_prob: 0.5})
        loss, acc = sess.run([Loss, accuracy], feed_dict = {x_image: data_x_in_resize, y_target: data_y_in, keep_prob: 1})
        val_acc = sess.run(accuracy, feed_dict = {x_image: validation_x_resize, y_target: data_y_onehot_val, keep_prob: 1})
        
        print("Epoch: " + str(i+1) + ' / ' + "Batch: " + str(j) + '/' + str(len_data) + ' / ' + 
              "Loss: " + str(loss) + ' / ' + "Training Accuracy: " + str(acc) + ' / ' + 
              "Validation Accuracy: " + str(val_acc))

```

## Testing

```python
# Testing

test_y_onehot = np.zeros([test_y.shape[0], num_label])
for i in range(test_y.shape[0]):
    test_y_onehot[i, int(test_y[i])] = 1

# Resize testset as image 
test_x_resize = np.reshape(test_x, (test_x.shape[0], img_size, img_size, 1))

test_result = output.eval(feed_dict = {x_image: test_x_resize, keep_prob: 1})

count_correct = 0
for i in range(test_result.shape[0]):
    prediction_y = np.argmax(test_result[i,:])
    
    if prediction_y == test_y[i]:
        count_correct += 1

test_acc = count_correct / test_result.shape[0]

print("Test Accuracy: " + str(test_acc))
```
